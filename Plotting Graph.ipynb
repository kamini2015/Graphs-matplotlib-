# adding label to graph

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(0,5,0.1)
y=x**3
y2=x**2
# plt.plot(x,y,color='black',marker='o')
plt.plot(x,y,color='black', linewidth=1,label="x^3")
plt.plot(x,y2,color='red', linewidth=1,label="x^2")
plt.ylabel("y")
plt.xlabel("x")
# using legend this will show our x^3 labels in the graph
plt.legend()
plt.title('Matplotlib')
plt.show()
```

```python
# Rainbow color bubble chart

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm

x = np.arange(10)
ys = [i+x+(i*x)**2 for i in range(10)]

colors = cm.rainbow(np.linspace(0, 1, len(ys)))
for y, c in zip(ys, colors):
    plt.scatter(x, y, color=c)
```

# Pie chart

```python
# to create the pie chart
import matplotlib.pyplot as plt
label = ["A","B", "C", "D"]
sizes=[3,4,6,2]
color = ["blue", "purple", "red", "pink"]

plt.pie(sizes,colors=color,labels=label)
# to make the size of pie equal
plt.axis("equal")
plt.show()
```

```python
# to create the pie chart with percentages
import matplotlib.pyplot as plt
label = ["A","B", "C", "D"]
sizes=[3,4,6,2]
color = ["lightblue", "coral", "pink", "seagreen"]

# 2 %% sign to show the icon in pie chart
# you can also pass the function in autopct

plt.pie(sizes,colors=color,labels=label,autopct='%0.1f%%')
# to make the size of pie equal
plt.axis("equal")
plt.show()
```

# passing function in autopct

```python
import matplotlib.pyplot as plt

# make the pie circular by setting the aspect ratio to 1
plt.figure(figsize=plt.figaspect(1))
values = [3, 12, 5, 8] 
labels = ['a', 'b', 'c', 'd'] 

def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{p:.2f}%  ({v:d})'.format(p=pct,v=val)
    return my_autopct

plt.pie(values, labels=labels, autopct=make_autopct(values))
plt.show()
```
